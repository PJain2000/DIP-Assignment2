### Implement an algorithm for low-pass filtering a grayscale image by moving a k Ã— k averaging filter of the form ones(k)/(k2). 

```python
%matplotlib inline

import cv2
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
i1 = './A2_resources/cameraman.png'

img1 = cv2.imread(i1, 0)
```

```python
def average(img, k):
    output = np.zeros_like(img)
    m = math.floor(float(k)/2)

    image_padded = np.zeros((img.shape[0] + 2*m, img.shape[1] + 2*m))   
    image_padded[m:-m, m:-m] = img

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            val = (image_padded[i:i+k,j:j+k]).sum()
            output[i,j] = val/(k*k)
            
    
    return output
```

```python
img2 = average(img1, 9)
```

```python
plt.subplot(121),plt.imshow(img1, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img2, cmap = 'gray')
plt.title('Average Filter'), plt.xticks([]), plt.yticks([])
plt.show()
```
